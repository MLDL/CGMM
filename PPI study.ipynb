```python
import torch
import torch_geometric
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
from torch_geometric.datasets import PPI
train = PPI('train')
```

```python
x_all = None
for i in range(len(train)):
    x = train[i].x
    
    if x_all is None:
        x_all = x
    else:
        x_all = torch.cat((x_all, x), axis=0)

print(x_all.shape)
```

```python
for i in range(50):
    plt.figure()
    plt.hist(x_all[:,i], bins=100)
    plt.yscale('log')
```

```python
from torch_geometric.utils import to_dense_adj
for i in range(len(train)):
    plt.figure()
    d = torch.sum(to_dense_adj(train[i].edge_index), axis=0)
    plt.hist(d)
    continue
```
